```python
import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
```

```python
plt.style.use('classic')

x = np.linspace(-10,10,101)
x
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.show()
```

```python
fig = plt.figure()

plt.plot(x,np.sin(x),"-")
plt.plot(x,np.cos(x),"--")
plt.plot(x,np.tan(x),"-")
```

```python
fig.savefig("my_grafico.pdf")
```

```python
fig.canvas.get_supported_filetypes()
```

```python
#Creando figura
plt.figure()

# row, columns, panel number
plt.subplot(2,2,1)
plt.plot(x,np.sin(x),"-")

plt.subplot(2,2,2)
plt.plot(x,np.cos(x),"--")

plt.subplot(2,2,3)
plt.plot(x,np.tan(x),"-")
```

```python
#subplots -> una tupla con dos posiciones
#ax -> array

fig,ax = plt.subplots(2,2)

ax[0][0].plot(x,np.sin(x),"-",color="#00CC99")
ax[0][1].plot(x,np.cos(x),"--",color="#FFEE88")
ax[1][1].plot(x,np.tan(x),"-.",color="#AA1155")
```

```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.sin(x),"8",color="#00CC99")
```

```python
#Añadiendo limites

plt.plot(x,np.sin(x),"8",color="#00CC99")

plt.xlim(-15,15)
plt.ylim(-1.5,1.5)
```

```python
# Atajo:
plt.plot(x,np.sin(x),"8",color="#00CC99")

plt.axis([-15,15,-1.5,1.5])
```

```python
plt.plot(x,np.sin(x),"8",color="#00CC99")

plt.axis('equal')
```

```python
plt.plot(x,np.sin(x),"8",color="#00CC99")

plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x,np.sin(x),"8",color="#00CC99",label="sin(x)")
plt.plot(x,np.cos(x),".",color="lightblue",label="cos(x)")

plt.axis("equal")
plt.xlim(-12,12)

plt.title("Datos")
plt.xlabel("x")

plt.legend()

```

```python
ax = plt.axes()

ax.plot(x,np.sin(x))

ax.set(xlim=(0,10),ylim=(-2,2),xlabel="x",ylabel="sin(x)",title="A simple plot")
```

# Scatter plots

```python
import matplotlib.pyplot as plt

plt.style.use("seaborn-whitegrid")
import numpy as np
```

```python
x = np.linspace(0,10,30)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y,'o',color="#648FFF")
```

```python
rng = np.random.RandomState(1)
fig = plt.figure()


for marker in ['-o', '-.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5),rng.rand(5),marker,label=f"marker={marker}")
plt.legend(numpoints=1)

plt.ylim(-0.2,1.2)
plt.xlim(-0.2,1.75)

fig.savefig("markers.jpg")
```

```python
plt.plot(x,y,"-s",color="#648FFF",linewidth=10,markersize=15,markeredgewidth=2,markeredgecolor="red")
```

```python
plt.scatter(x,y,marker="o")
```

```python
rng = np.random.RandomState(1)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar()
```
