```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-white")

data = np.random.randn(1000)
data
```

```python
plt.hist(data,bins=30,label="Data",rwidth=0.5,align="mid",range=(-4,4),density=True,alpha=0.5,histtype="barstacked",color="red",edgecolor="none")
```

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)
```

```python
plt.hist(x1,alpha=0.3,bins=40)
plt.hist(x2,alpha=0.3,bins=40)
plt.hist(x3,alpha=0.3,bins=40)
```

```python
counts,bin_edges = np.histogram(x1,bins=5)

bin_edges
```
