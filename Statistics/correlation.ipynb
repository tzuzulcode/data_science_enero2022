```python
import pandas as pd
import seaborn as seaborn
```

```python
sp500 = pd.read_csv("../dataset/sp500_data.csv")
sectors = pd.read_csv("../dataset/sp500_sectors.csv")
```

```python
sp500.tail(10)
```

```python
it_companies = sp500.loc[sp500["Unnamed: 0"]>"2012-07-01",sectors[sectors["sector"]=="information_technology"]["symbol"]]
filtered_companies = it_companies.loc[:,"ADS":"IBM"]
filtered_companies
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
it_companies.corr()
```

```python
sns.heatmap(it_companies.corr(),vmin=-1,vmax=1,annot=True,figsize=(10,10),cmap=sns.diverging_palette(10, 240, n=12,as_cmap=True))
```

```python
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(20, 20))
    sns.heatmap(filtered_companies.corr(),vmin=-1,vmax=1,cmap=sns.diverging_palette(10, 240, n=12,as_cmap=True))
```

# Scatterplots
Muestran la relacion entre dos variables

```python
ax = sp500.loc[sp500["Unnamed: 0"]>"2012-07-01"].plot.scatter(x="MSFT",y="IBM",figsize=(10,10))
ax.set_xlabel("Microsoft")
ax.set_ylabel("IBM")

ax.axhline(0,color="grey",lw=1)
ax.axvline(0,color="grey",lw=1)
```
