```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../../dataset/loans_income.csv")
data = data.income
data
```

```python
confidence = 95
left_confidence = ((100-confidence)/2)/100
right_confidence = (100 - (100-confidence)/2)/100
print(left_confidence)
print(right_confidence)
```

```python
values = [np.random.choice(data,size=20,replace=True).mean() for i in range(1000)]
values
```

```python
confidence_interval = np.percentile(values,[left_confidence,right_confidence])

confidence_interval
```

```python
mean = np.array(values).mean()
mean
```
