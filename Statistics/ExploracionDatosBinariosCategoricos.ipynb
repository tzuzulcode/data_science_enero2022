# Expected value (Valor esperado)
Es informacin en las que las categorias representan o pueden ser mapeadas a valores discretos en la misma escala

El valor esperado: Una forma de media ponderada en el cual los pesos son las probabilidades

```python
# Calcular valor esperado de los servicios en la nube:
# 300 -> 0.05
# 50 -> 0.15
# 0 -> 0.8

expected_value = (0.05*300) + (0.15*50) +(0 * 0.8)
expected_value
```
