```python
import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
```

```python
plt.style.use('classic')

x = np.linspace(-10,10,101)
x
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.show()
```

```python
fig = plt.figure()

plt.plot(x,np.sin(x),"-")
plt.plot(x,np.cos(x),"--")
plt.plot(x,np.tan(x),"-")
```

```python
fig.savefig("my_grafico.pdf")
```

```python
fig.canvas.get_supported_filetypes()
```

```python
#Creando figura
plt.figure()

# row, columns, panel number
plt.subplot(2,2,1)
plt.plot(x,np.sin(x),"-")

plt.subplot(2,2,2)
plt.plot(x,np.cos(x),"--")

plt.subplot(2,2,3)
plt.plot(x,np.tan(x),"-")
```

```python
#subplots -> una tupla con dos posiciones
#ax -> array

fig,ax = plt.subplots(2,2)

ax[0][0].plot(x,np.sin(x),"-",color="#00CC99")
ax[0][1].plot(x,np.cos(x),"--",color="#FFEE88")
ax[1][1].plot(x,np.tan(x),"-.",color="#AA1155")
```
